## Demo: How to work with database


Download database file.


```python
# !wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("survey_data.sqlite") # open a database connection
```

Import needed modules.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
query = '''
SELECT * 
    FROM master
'''
df = pd.read_sql_query(query, conn)

sns.set(font_scale=1.5, rc={'figure.figsize':(15, 6)})
sns.displot(df.ConvertedComp, edgecolor='black', kde=True, color='#882855', height=5, aspect=2.5, alpha=.6, bins=40).set(xlabel='Salary (USD)')
plt.text(x=.725*10**6, y=5500, s='Salary Distribution', fontsize=25)
plt.text(x=.75*10**6, y=5200, s='The width of each bin is 50K', fontsize=15);
```

### Box Plots


Plot a box plot of `Age.`


```python
sns.boxplot(data = df.Age, color='#882855', orient='h').set(xlabel='Age', title='Five Number Summary of Age');
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
sns.scatterplot(x=df.Age, y=df.WorkWeekHrs, data=df, hue='WorkWeekHrs').set(title='The Relationship Between Age and WorkWeekHrs');
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
sns.scatterplot(x=df.WorkWeekHrs, y=df.CodeRevHrs, data=df, size='Age', sizes=(0, 500), hue='Age').set(title='Relationship of CodeReviewHrs with WorkWeekHrs');
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Extract the data from the database
query = '''
SELECT DatabaseDesireNextYear, Count(*) as Count
    FROM DatabaseDesireNextYear
    GROUP BY DatabaseDesireNextYear
    ORDER BY count DESC limit 5
'''
db_df = pd.read_sql_query(query, conn)
db_df.set_index('DatabaseDesireNextYear', inplace=True)
db_df
```

```python
palette_color = sns.color_palette("rocket_r")
explode = [.1, 0, 0, 0, 0]
plt.pie(db_df.Count, labels=db_df.index.values, autopct='%.1f%%', shadow=True, colors=palette_color, textprops={'color':'#000000'}, explode=explode);
plt.title('Top Databases That Respondents Want To Learn Next Year')
plt.axis('equal');
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
age_3035 = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median().loc[30:35, ]
age_3035.index = age_3035.index.astype(int)

age_3035.plot(kind='bar', stacked=True, color=['#882855', '#ff7f50'], ylabel='Median of Work Week Hrs')
plt.text(x=.5, y=50, s='Relationship of age to time spent reviewing code from WorkWeekHrs')
plt.text(x=1.5, y=47, s='The ages are between 30 to 35', fontsize=15);
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
age_4560 = df.groupby('Age')[['ConvertedComp']].median().loc[45:60, ]
age_4560.rename(columns={'ConvertedComp':'Salary (USD)'}, inplace=True)

sns.lineplot(data=age_4560, palette='rocket')
plt.text(x=49, y=13.9*10**4, s='Relationship of Median Salary With Age')
plt.text(x=50, y=13.6*10**4, s='The ages are between 45 to 60', fontsize=15);
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`

```python
dt = df.MainBranch.value_counts().to_frame()
dt.index = ['Primarily Developer', 'Not Primarily Developer']
palette_color = sns.color_palette('rocket_r')
sns.barplot(data=dt, x='MainBranch', y=dt.index, palette=palette_color, orient='h').set(title='Distribution of Respondents\' Profession',xlabel='Count');
```
